```python
import os
import pandas as pd

# asos 폴더 내의 모든 파일의 경로를 가져옵니다.
folder_path = 'dust'
all_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.csv')]

# 각 파일의 행 개수를 저장하는 딕셔너리
row_counts = {}

total_rows = 0  # 모든 파일에서 행의 총합을 저장할 변수

for file in all_files:
    with open(file, 'r', encoding='cp949') as f:
        df = pd.read_csv(f)
    # 파일 이름과 해당 파일의 행 개수를 딕셔너리에 저장
        row_counts[file] = len(df)
        total_rows += len(df)

# 결과를 출력
for file, count in row_counts.items():
    print(f"{file}: {count} rows")

print("\nTotal rows across all files:", total_rows)
```

```python
# asos 폴더 내의 모든 파일의 경로를 가져옵니다.
folder_path = 'dust'
all_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.csv')]

list_of_dfs = []
for file in all_files:
    with open(file, 'r', encoding='cp949') as f:
        df = pd.read_csv(f)
        list_of_dfs.append(df)

# 모든 DataFrame을 하나로 결합합니다.
combined_df = pd.concat(list_of_dfs, ignore_index=True)

# 결과를 출력합니다.
print(combined_df.head())

# 필요에 따라 합쳐진 DataFrame을 CSV 파일로 저장할 수 있습니다.
combined_df.to_csv('combined_dust.csv', index=False, encoding='cp949')
```

```python
# 지우고 싶은 범위를 설정
start_date = '2013-01-01 00:00:00'
end_date = '2013-03-19 00:00:00'

# 해당 범위에 있는 데이터를 필터링
mask = (dust['시간'] >= start_date) & (dust['시간'] <= end_date)

# 해당 범위의 데이터 삭제
dust = dust[~mask]
print(dust.head(50))
```

```python
len(dust)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
import pandas as pd

# '시간' 칼럼을 datetime 유형으로 변환
dust['시간'] = pd.to_datetime(dust['시간'])

# 2013년부터 2022년까지의 완전한 날짜 및 시간 범위 생성
full_date_range = pd.date_range(start='2013-03-19 01:00', end='2022-12-31 23:00', freq='H')

# 누락된 날짜 및 시간 찾기
missing_dates = full_date_range[~full_date_range.isin(dust['시간'])]

print(missing_dates)
print(len(missing_dates))
```

```python
duplicated_dates = dust[dust['시간'].duplicated()]['시간']
if duplicated_dates.shape[0] > 0:
    print("There are duplicated dates in the '시간' column.")
    print(duplicated_dates)
else:
    print("There are no duplicated dates in the '시간' column.")
```

```python
# NaT 값들이 있는 행 출력
print(dust[dust['시간'].isna()])
```

```python
# '날짜' 컬럼에 값이 있고 '시간' 컬럼에 NaT 값이 있는 행을 찾아 이동
if '일시' in dust.columns and '시간' in dust.columns:
    mask = dust['시간'].isna() & ~dust['일시'].isna()
    dust.loc[mask, '시간'] = dust.loc[mask, '일시']
    dust.drop(columns=['일시'], inplace=True)

# 변경된 데이터셋을 저장
#dust.to_csv('merged0110.csv', index=False)
```
