```python
with open('temp.csv', 'r', encoding='cp949') as f:
    data = pd.read_csv(f)
data.head()
```

```python
len(data)
```

```python
# 컬럼 이름 변경
column_translations = {
    '일시': 'DateTime',
    'month': 'Month',
    'day': 'Day',
    'hour': 'Hour',
    '기온(°C)': 'Temperature(C)',
    '강수량(mm)': 'Precipitation(mm)',
    '습도(%)': 'Humidity(%)',
    '증기압(hPa)': 'VaporPressure(hPa)',
    '이슬점온도(°C)': 'DewPointTemperature(C)',
    '일조(hr)': 'Sunshine(hr)',
    '일사(MJ/m2)': 'SolarRadiation(MJ/m2)',
    '적설(cm)': 'Snowfall(cm)',
    '3시간신적설(cm)': 'SnowfallLast3Hours(cm)',
    '전운량(10분위)': 'TotalCloudCoverage(1/10)',
    '중하층운량(10분위)': 'MidLowCloudCoverage(1/10)',
    '운형(운형약어)': 'CloudForm',
    '최저운고(100m)': 'LowestCloudHeight(100m)',
    '시정(10m)': 'Visibility(10m)',
    '지면상태(지면상태코드)': 'GroundState(code)',
    '현상번호(국내식)': 'PhenomenonNumber',
    '최저운고(100m )':'LowCloud(100m)'
}

data_translated_columns = data.rename(columns=column_translations)

# 변경된 데이터의 처음 5행 확인
data_translated_columns.head()
```
