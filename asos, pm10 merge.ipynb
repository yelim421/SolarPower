```python
import pandas as pd
import os
```

```python
# asos 폴더 내의 모든 파일의 경로를 가져옵니다.
folder_path = 'asos'
all_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.csv')]

# 각 파일의 열 개수를 저장하는 딕셔너리
column_counts = {}


for file in all_files:
    with open(file, 'r', encoding='cp949') as f:
        df = pd.read_csv(f)
    # 파일 이름과 해당 파일의 열 개수를 딕셔너리에 저장
        column_counts[file] = len(df.columns)

# 결과를 출력
for file, count in column_counts.items():
    print(f"{file}: {count} columns")

# 모든 파일의 열 개수가 동일한지 확인
if len(set(column_counts.values())) == 1:
    print("All CSV files have the same number of columns.")
else:
    print("CSV files have different numbers of columns.")
```

```python
# asos 폴더 내의 모든 파일의 경로를 가져옵니다.
folder_path = 'asos'
all_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.csv')]

list_of_dfs = []
for file in all_files:
    with open(file, 'r', encoding='cp949') as f:
        df = pd.read_csv(f)
        list_of_dfs.append(df)

# 모든 DataFrame을 하나로 결합합니다.
combined_df = pd.concat(list_of_dfs, ignore_index=True)

# 결과를 출력합니다.
print(combined_df.head())

# 필요에 따라 합쳐진 DataFrame을 CSV 파일로 저장할 수 있습니다.
combined_df.to_csv('combined_asos.csv', index=False, encoding='cp949')
```
