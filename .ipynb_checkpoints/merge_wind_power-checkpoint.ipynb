```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('test_data_nan.csv', index_col='DateTime')
wind = pd.read_csv('wind_speed.csv')
```

```python
# 'DateTime' 인덱스와 'Date'를 기준으로 병합
merged_df = train.merge(wind, left_on=train.index, right_on='Date', how='left')

# 'Wind_Speed' 값을 'wind_power'에 복사
merged_df['wind_power'] = merged_df['Wind_Speed']

# 인덱스를 'DateTime'으로 설정
merged_df.set_index('DateTime', inplace=True)

# 불필요한 'Date' 컬럼 제거
final_train = merged_df.drop(columns=['Date'])

final_train.head()
```

```python
merged_df = train.merge(wind, left_on=train.index, right_on='Date', how='left')
merged_df.head()
```

```python
merged_df.set_index('Date', inplace=True)
merged_df.head(50)
```

```python
merged_df.head(50)
```

```python
# 기초 통계량 확인
print(merged_df['Wind_Speed'].describe())

# 결측값 확인
num_nan = merged_df['Wind_Speed'].isna().sum()
print(f"\nNumber of NaN values: {num_nan}")

# 0 값 확인
num_zeros = (merged_df['Wind_Speed'] == 0).sum()
print(f"Number of 0 values: {num_zeros}")

# 히스토그램 그리기
merged_df['Wind_Speed'].hist(bins=30, figsize=(10, 6))
plt.title('Distribution of Wind_Speed')
plt.xlabel('Wind_Speed')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
merged_df.to_csv('test_data_nan_wind.csv', index=True)
```
