```python
import pandas as pd
```

```python
import xarray as xr
DS = xr.open_dataset("adaptor.mars.internal-1691650570.6133935-15681-11-856ebbea-d151-4fc3-9fd9-35d3137cc735.nc")
DS.to_dataframe().to_csv("2mt_2.csv")
```

```python
with open('2mt_1.csv', 'r') as f:
    df = pd.read_csv(f)
df.head(15)
```

```python
with open('2mt_2.csv', 'r') as f:
    df2 = pd.read_csv(f)
df2.head(15)
```

```python
combined_df = pd.concat([df, df2], ignore_index=True)

# 필요한 경우 연결된 데이터프레임을 새로운 csv 파일로 저장합니다.
combined_df.to_csv('combined_2mt.csv', index=False)
```

```python
with open('combined_2mt.csv', 'r') as f:
    df = pd.read_csv(f)
df.head(15)
```

```python
# 지우고 싶은 범위를 설정
start_date = '2013-01-01 00:00:00'
end_date = '2013-03-18 23:00:00'

# 해당 범위에 있는 데이터를 필터링
mask = (df['일시'] >= start_date) & (df['일시'] <= end_date)

# 해당 범위의 데이터 삭제
df = df[~mask]
print(df.head(50))
```
