```python
import pandas as pd
import os
```

```python
# asos 폴더 내의 모든 파일의 경로를 가져옵니다.
folder_path = 'asos'
all_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.csv')]

# 각 파일의 열 개수를 저장하는 딕셔너리
column_counts = {}


for file in all_files:
    with open(file, 'r', encoding='cp949') as f:
        df = pd.read_csv(f)
    # 파일 이름과 해당 파일의 열 개수를 딕셔너리에 저장
        column_counts[file] = len(df.columns)

# 결과를 출력
for file, count in column_counts.items():
    print(f"{file}: {count} columns")

# 모든 파일의 열 개수가 동일한지 확인
if len(set(column_counts.values())) == 1:
    print("All CSV files have the same number of columns.")
else:
    print("CSV files have different numbers of columns.")
```

```python
# asos 폴더 내의 모든 파일의 경로를 가져옵니다.
folder_path = 'asos'
all_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.csv')]

list_of_dfs = []
for file in all_files:
    with open(file, 'r', encoding='cp949') as f:
        df = pd.read_csv(f)
        list_of_dfs.append(df)

# 모든 DataFrame을 하나로 결합합니다.
combined_df = pd.concat(list_of_dfs, ignore_index=True)

# 결과를 출력합니다.
print(combined_df.head())

# 필요에 따라 합쳐진 DataFrame을 CSV 파일로 저장할 수 있습니다.
combined_df.to_csv('combined_asos.csv', index=False, encoding='cp949')
```

```python
import pandas as pd

# 두 csv 파일 읽기
with open('asos_before_merge.csv', 'r', encoding='cp949') as f:
    df1 = pd.read_csv(f)
df2 = pd.read_csv('power_before_merge.csv')


df3['일시'] = pd.to_datetime(df3['일시'])
df2['date'] = pd.to_datetime(df2['date'])

# datetime을 원하는 문자열 형식으로 변경
df3['일시'] = df3['일시'].dt.strftime('%Y-%m-%d %H:00')
df2['date'] = df2['date'].dt.strftime('%Y-%m-%d %H:00')

# "일시" 열의 값들을 집합으로 변환
set1 = set(df3['일시'])
set2 = set(df2['date'])

# 한쪽에만 있는 값을 찾기
only_in_df1 = set1 - set2
only_in_df2 = set2 - set1

print("Only in file1:", only_in_df1)
print("Only in file2:", only_in_df2)
```

```python
# df1의 '일시' 포맷을 'YYYY-MM-DD HH:MM'으로 변환
df1['일시'] = pd.to_datetime(df1['일시']).dt.strftime('%Y-%m-%d %H:%M')

# df2의 '일시' 포맷도 동일하게 변환
df2['date'] = pd.to_datetime(df2['date'], errors='coerce').dt.strftime('%Y-%m-%d %H:%M')

# '일시' 열을 기준으로 두 데이터프레임을 정렬
sorted_df1 = df1.sort_values(by='일시')
sorted_df2 = df2.sort_values(by='date')


sorted_df1 = sorted_df1.reset_index(drop=True)
sorted_df2 = sorted_df2.reset_index(drop=True)

# 가능한 길이로 비교 (더 짧은 데이터프레임의 길이)
min_length = min(len(sorted_df1), len(sorted_df2))

# 해당 길이만큼만 비교
diff_mask = sorted_df1['일시'][:min_length] != sorted_df2['date'][:min_length]
print("\nDifferences:")
print(pd.concat([sorted_df1['일시'][diff_mask], sorted_df2['date'][diff_mask]], axis=1))
```

```python
dates_df1 = sorted_df1['일시'].tolist()
dates_df2 = sorted_df2['date'].tolist()

# 두 리스트에서의 차이점 찾기
only_in_df1 = set(dates_df1) - set(dates_df2)
only_in_df2 = set(dates_df2) - set(dates_df1)

print("Only in df1:", only_in_df1)
print("Only in df2:", only_in_df2)
```

```python
# df1과 df2의 '일시' 열에서 중복 확인
duplicates_df1 = df1[df1.duplicated(subset=['일시'], keep=False)]
duplicates_df2 = df2[df2.duplicated(subset=['date'], keep=False)]

print("Duplicates in df1:")
print(duplicates_df1['일시'])

print("\nDuplicates in df2:")
print(duplicates_df2['date'])

# '일시' 열을 기준으로 두 데이터프레임을 정렬
sorted_df1 = df1.sort_values(by='일시')
sorted_df2 = df2.sort_values(by='date')

# 두 데이터프레임의 '일시' 열 값이 다른 부분을 찾아 출력
diff_mask = sorted_df1['일시'] != sorted_df2['date']
print("\nDifferences:")
print(pd.concat([sorted_df1['일시'][diff_mask], sorted_df2['date'][diff_mask]], axis=1))
```
