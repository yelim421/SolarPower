```python

# 2. 각 CloudForm에서 구름 종류 추출
cloud_types = set()

for form in df['CloudForm'].dropna().unique():
    cloud_types.update(re.findall(r'[A-Z][a-z]', form))

cloud_types_list = sorted(list(cloud_types))

# 3. 원-핫 인코딩 진행
for cloud_type in cloud_types_list:
    df[f'Cloud_{cloud_type}'] = df['CloudForm'].str.contains(cloud_type).fillna(0).astype(int)

# 4. NaN 값에 대한 열 추가
df['CloudForm_NaN'] = df['CloudForm'].isna().astype(int)

# 결과 확인
df[['CloudForm'] + [f'Cloud_{cloud_type}' for cloud_type in cloud_types_list] + ['CloudForm_NaN']].head()
```
