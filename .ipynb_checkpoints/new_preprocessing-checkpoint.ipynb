```python
df = df.drop(columns = 'total')
df.head()
```

```python
df['year'] = pd.to_datetime(df['date']).dt.year
df['month'] = pd.to_datetime(df['date']).dt.month
df['day'] = pd.to_datetime(df['date']).dt.day
df = df.drop(columns='date')
df.head()
```

```python
df_melted = df.melt(id_vars=['year', 'month', 'day'], 
                   var_name='hour', 
                   value_name='power')
df_melted.head()
```

```python
# 'year', 'month', 'day', 'hour' 순으로 오름차순 정렬
df_melted = df_melted.sort_values(by=['year', 'month', 'day', 'hour'])

pd.set_option('display.max_rows', None)

# 결과 확인
print(df_melted.head(200))
```

```python
zero_power_counts = df_melted[df_melted['power'] != 0.0].groupby('hour').size()

print(zero_power_counts)
```

```python
zero_power_counts = df_melted[df_melted['power'] == 0.0].groupby('hour').size()

print(zero_power_counts)
```
