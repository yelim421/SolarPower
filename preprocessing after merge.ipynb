```python
def summary(df):
    sum = pd.DataFrame(df.dtypes, columns=['dtypes'])
    sum['missing#'] = df.isna().sum().values*100
    sum['missing%'] = (df.isna().sum().values*100)/len(df)
    sum['uniques'] = df.nunique().values
    sum['count'] = df.count().values
    #sum['skew'] = df.skew().values
    desc = pd.DataFrame(df.describe().T)
    sum['min'] = desc['min']
    sum['max'] = desc['max']
    sum['mean'] = desc['mean']
    return sum

summary(df).style.background_gradient(cmap='Blues')
```

```python
# '일시'를 datetime 객체로 변환
df['일시'] = pd.to_datetime(df['일시'])

# datetime 객체에서 각 부분을 추출하여 새로운 열에 저장
df['year'] = df['일시'].dt.year
df['month'] = df['일시'].dt.month
df['day'] = df['일시'].dt.day
df['hour'] = df['일시'].dt.hour

# 새로운 열들을 데이터프레임의 맨 앞에 배치
df = df[['year', 'month', 'day', 'hour'] + [col for col in df if col not in ['year', 'month', 'day', 'hour']]]
df.head()
```

```python
# line plot for emision 
sns.lineplot(data = df, x = 'month', y = 'power', errorbar = None)
    
plt.title('power Over month', fontsize = 24, fontweight = 'bold')
plt.show()
```

```python
# 년도별로 색깔을 다르게 하여 라인 플롯을 그립니다.
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='month', y='power', hue='year', palette="tab10", marker="o")

plt.title('power Over month by Year', fontsize=24, fontweight='bold')
plt.show()
```

```python
# 월별 'power'의 누적 합계를 계산합니다.
df_cumulative = df.groupby('month')['power'].sum().cumsum().reset_index()

# 누적 합계를 사용하여 라인 플롯을 그립니다.
sns.lineplot(data=df_cumulative, x='month', y='power', errorbar=None)

plt.title('Cumulative power Over month', fontsize=24, fontweight='bold')
plt.show()
```

```python
# 년도별, 월별 'power'의 누적 합계를 계산합니다.
df_cumulative = df.groupby(['year', 'month'])['power'].sum().groupby(level=0).cumsum().reset_index()

# 누적 합계를 사용하여 라인 플롯을 그립니다.
plt.figure(figsize=(12,6))
sns.lineplot(data=df_cumulative, x='month', y='power', hue='year', palette="tab10", marker="o")

plt.title('Yearly Cumulative power Over Month', fontsize=24, fontweight='bold')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

num = df.columns.tolist()[1:-1] # 적절하게 필요한 열들을 선택하세요

fig, axes = plt.subplots(len(num), 2, figsize=(16, len(num) * 4.2), 
                         gridspec_kw={'hspace': 0.35, 'wspace': 0.3, 'width_ratios': [0.80, 0.20]});

for i, col in enumerate(num):
    ax = axes[i, 0]
    sns.kdeplot(data=df[col], ax=ax, linewidth=2.1)
    ax.set_title(f"\n{col}", fontsize=9, fontweight='bold')
    ax.grid(visible=True, which='both', linestyle='--', color='lightgrey', linewidth=0.75)
    ax.set(xlabel='', ylabel='')

    ax = axes[i, 1]
    sns.boxplot(data=df[col], y=col, width=0.25, saturation=0.90, linewidth=0.90, fliersize=2.25, color='#037d97', ax=ax)
    ax.set(xlabel='', ylabel='')
    ax.set_title(f"Train", fontsize=9, fontweight='bold')

plt.suptitle(f"\nDistribution analysis- continuous columns\n", fontsize=12, fontweight='bold', y=0.89, x=0.57)
plt.tight_layout()
plt.show()
```
