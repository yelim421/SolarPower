```python
import pandas as pd
import os

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
with open('dust.csv', 'r') as f:
    aerosol = pd.read_csv(f)
aerosol.head()
```

```python
with open('asos_power_merge_ymdh.csv', 'r', encoding='cp949') as f:
    asos = pd.read_csv(f)
asos.head()
```

```python
with open('combined_2mt.csv', 'r') as f:
    t2m = pd.read_csv(f)
print(len(t2m))
t2m.head()
```

```python
# 각 데이터프레임의 인덱스를 시간으로 설정합니다.
aerosol.set_index('time_copy', inplace=True)
asos.set_index('일시', inplace=True)
t2m.set_index('time', inplace=True)

# 먼저 aerosol과 asos를 합칩니다.
merged_df = aerosol.merge(asos, left_index=True, right_index=True)

# 그 다음, 이전에 합친 데이터프레임과 t2m를 합칩니다.
final_df = merged_df.merge(t2m, left_index=True, right_index=True)
```

```python
final_df.head()
```

```python
final_df.rename(columns={
    '기온(°C)': 'Temp(C)',
    '강수량(mm)': 'Prec(mm)',
    '습도(%)': 'Humidity(%)',
    '이슬점온도(°C)': 'Dew_Point(C)',
    '일사(MJ/m2)': 'Solar_Radiation(MJ/m2)',
    '적설(cm)': 'Snowfall(cm)',
    '3시간신적설(cm)': '3hr_New_Snowfall(cm)',
    '전운량(10분위)': 'Cloud_Cover(1/10)',
    '중하층운량(10분위)': 'Middle_Low_Cloud_Cover(1/10)',
    '최저운고(100m )': 'Lowest_Cloud_Height(100m)',
    '시정(10m)': 'Visibility(10m)',
    '30cm 지중온도(°C)': '30cm_Ground_Temperature(C)'
}, inplace=True)
final_df.head()
```

```python
final_df.to_csv('all.csv', index=False)
```

```python
final_df.head(100)
```

```python
filtered_data = final_df[final_df['power'] != 0.0]
filtered_data.to_csv('all_chuli_power.csv', index=False)
filtered_data.head(100)
```

```python
columns_to_fill_zero = ['Snowfall(cm)', '3hr_New_Snowfall(cm)', 'Prec(mm)']
for column in columns_to_fill_zero:
    filtered_data[column].fillna(0, inplace=True)
filtered_data.head(50)
```

```python
# 'time_copy'를 DatetimeIndex로 설정
filtered_data.index = pd.to_datetime(filtered_data.index)
```

```python
filtered_data['Lowest_Cloud_Height(100m)'].interpolate(method='time', inplace=True)

# 필요하다면 인덱스를 reset 할 수 있습니다.
# data.reset_index(inplace=True)
```

```python
filtered_data['Cloud_Cover(1/10)'] = filtered_data['Cloud_Cover(1/10)'].interpolate(method='time')
filtered_data['Middle_Low_Cloud_Cover(1/10)'] = filtered_data['Middle_Low_Cloud_Cover(1/10)'].interpolate(method='time')
```

```python
filtered_data.head(50)
```

```python
nan_locations = filtered_data.isnull()

# 각 셀의 위치를 출력한다.
for column in nan_locations.columns:
    for index, value in nan_locations[column].items():
        if value:
            print(f"NaN 값이 위치한 셀: 행 {index}, 열 {column}")
```

```python
filtered_data['30cm_Ground_Temperature(C)'] = filtered_data['30cm_Ground_Temperature(C)'].interpolate(method='time')
filtered_data['Visibility(10m)'] = filtered_data['Visibility(10m)'].interpolate(method='time')
filtered_data['Solar_Radiation(MJ/m2)'] = filtered_data['Solar_Radiation(MJ/m2)'].interpolate(method='time')
filtered_data['Dew_Point(C)'] = filtered_data['Dew_Point(C)'].interpolate(method='time')
filtered_data['Humidity(%)'] = filtered_data['Humidity(%)'].interpolate(method='time')
filtered_data['Temp(C)'] = filtered_data['Temp(C)'].interpolate(method='time')
```

```python
filtered_data.to_csv('done_nan.csv', index=False)
filtered_data.head(100)
```
