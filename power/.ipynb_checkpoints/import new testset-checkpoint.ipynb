```python
import pandas as pd
```

```python
test1 = pd.read_csv('한국서부발전(주)_신재생에너지발전량_20220228.csv', encoding='cp949')
test2 = pd.read_csv('한국서부발전(주)_태양광 발전 현황_20220731.csv', encoding='cp949')
```

```python
df_melted = test1.melt(id_vars=['년월일'], value_vars=[f'{i:02d}시' for i in range(1, 6)], var_name='Hour', value_name='power')
```

```python
# 발전기명에 대한 고유값(종류)를 얻기
unique_generators = test1['발전기명'].unique()

# 각 발전기명별 행의 수를 얻기
generator_counts = test1['발전기명'].value_counts()

print("발전기 종류:", unique_generators)
print("\n각 발전기별 행의 수:\n", generator_counts)
```

```python
# 발전기명에 대한 고유값(종류)를 얻기
unique_generators = test2['발전기명'].unique()

# 각 발전기명별 행의 수를 얻기
generator_counts = test2['발전기명'].value_counts()

print("발전기 종류:", unique_generators)
print("\n각 발전기별 행의 수:\n", generator_counts)
```
