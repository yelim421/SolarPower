```python
train = pd.read_csv('train_data_nan_wind.csv', index_col='Date')
train.fillna(-1, inplace=True)
train.head()
```

```python
import numpy as np
# 'Snowfall(cm)'의 구간별 평균 'power' 값 계산
snowfall_bins = [-np.inf, 0.1, 0.3, np.inf]
snowfall_labels = ["Low", "Medium", "High"]
train['Snowfall Impact'] = pd.cut(abs(train['Snowfall(cm)']), bins=snowfall_bins, labels=snowfall_labels)
avg_power_snowfall = train.groupby('Snowfall Impact')['power'].mean()

# 'Precipitation(mm)'의 구간별 평균 'power' 값 계산
precipitation_bins = [-np.inf, 0.1, 0.3, np.inf]
precipitation_labels = ["Low", "Medium", "High"]
train['Precipitation Impact'] = pd.cut(abs(train['Precipitation(mm)']), bins=precipitation_bins, labels=precipitation_labels)
avg_power_precipitation = train.groupby('Precipitation Impact')['power'].mean()

avg_power_snowfall, avg_power_precipitation
```

```python
# Correcting the variable names based on the previously defined thresholds

def classify_snowfall(value):
    if value <= 0.1:
        return 1
    elif value <= 0.3:
        return 2
    else:
        return 3


train['Snowfall_Class'] = train['Snowfall(cm)'].apply(classify_snowfall)
train['Precipitation_Class'] = train['Precipitation(mm)'].apply(classify_snowfall)

train.head()
```
